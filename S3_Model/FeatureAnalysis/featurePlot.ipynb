```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = r'C:\Users\Sir\OneDrive\Univeristy\HEA Analysis\4K06\Code\Data Folder\Processed 0922\Tbreak_Yield.csv'
df = pd.read_csv(file_path)

# Specify the columns that need to be plotted
columns_to_plot = [
    "VEC", "ElectDiff", "Atomic Size Diff", "Mixing Enthalpy", "Mixing Entropy",
    "AVG Shear Modulus", "AVG Bulk Modulus", "AVG Possion Ratio", "Itinerate Electron",
    "Cohesive Energy", "Gamma", "Local Size Mismatch", "Local Shear Modulus Mismatch",
    "Local Electronegativity Mismatch", "Avg Melting Pt", "Cat"
]

# Filter the DataFrame for the required columns
filtered_df = df[columns_to_plot]

# Convert all relevant columns to numeric, forcing errors to NaN
filtered_df[filtered_df.columns[:-1]] = filtered_df[filtered_df.columns[:-1]].apply(pd.to_numeric, errors='coerce')

# Drop rows with missing data
filtered_df_clean = filtered_df.dropna()

# Define function to generate boxplots
def plot_box_distributions(df, category_col="Cat"):
    for column in df.columns[:-1]:  # Exclude the 'Cat' column
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=category_col, y=column, data=df)
        plt.title(f'Boxplot of {column} by Category')
        plt.xlabel('Category')
        plt.ylabel(column)
        plt.savefig(fr'C:\Users\Sir\OneDrive\Univeristy\HEA Analysis\4K06\Code\Data Folder\Processed 0922\FeatureBox\{column}.png')
        plt.show()

# Generate the boxplots
plot_box_distributions(filtered_df_clean)
```
